<h4 style="color: #008000;">Questão 2</h4>

Sendo:
<p>
    \( g(t) = \cos(\omega_1 t)\)
</p>
<p>
    \( h(t) = \cos(\omega_1 t + \theta) \)
</p>
Determine: 
<p>
    \( s(t) = g(t)h(t)\)
</p>
Escolha três valores para θ e esboce os sinais resultantes.

<p>
a) \( \theta = \pi \)
</p>

```python
t = np.linspace(0, 5, 1000) # Cria o eixo do tempo

w = 2
theta = np.pi

g = np.cos(w * t)
h = np.cos(w* t + theta)
s = g*h

plt.plot(t, s)
plt.yticks(np.arange(-1, 0.1, 0.1))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<p>
b) \( \theta = \frac{\pi}{2} \)
</p>

```python
theta = np.pi/2

g = np.cos(w * t)
h = np.cos(w* t + theta)
s = g*h

plt.plot(t, s)
plt.yticks(np.arange(-1, 1, 0.1))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<p>
c) \( \theta = 2\pi \)
</p>

```python
theta = 2 * np.pi

g = np.cos(w * t)
h = np.cos(w* t + theta)
s = g*h

plt.plot(t, s)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<h4 style="color: #008000;">Questão 10</h4>

Desenhe os sinais:

<p style="text-align: center; font-size: 16px;">
    \( x_1(t) = \sum_{k=-\infty}^{\infty} \delta(t - kT_1) \)
</p>
<p style="text-align: center; font-size: 16px;">
    \( x_2(t) = 2\sum_{k=-\infty}^{\infty} \delta(t - mT_2) \)
</p>

```python
# período dos impulsos
T1 = 1

# eixo do tempo
t = np.arange(-5, 6, 1)

# cria os impulsos
x = np.zeros_like(t)

# coloca 1 nos múltiplos de T1
x[t % T1 == 0] = 1

# gráfico
plt.stem(t, x)

plt.xlabel('t')
plt.xticks(np.arange(-5, 6, 1))
plt.ylabel('x₁(t)')

plt.grid()
plt.show()
```

```python
# período dos impulsos
T2 = 1

# eixo do tempo
t = np.arange(-5, 6, 1)

# vetor zerado
x = np.zeros_like(t)

# impulsos com amplitude 2
x[t % T2 == 0] = 2

# gráfico
plt.stem(t, x)

plt.xlabel('t')
plt.xticks(np.arange(-5, 6, 1))
plt.ylabel('l2(t)')

plt.grid()
plt.show()
```

<h4 style="color: #008000;">Questão 26</h4>
Seja:

<p style="text-align: center; font-size: 16px;">
    \( x(t) = \cos(\omega t + \phi) \)
</p>

```python
# vetor de tempo
t = np.linspace(-6, 6, 1000)

omega = 3       # Omega variável
phi = np.pi / 4

# sinal original
x = np.cos(omega * t + phi)

# sinal espelhado no tempo
x_neg = np.cos(omega * (-t) + phi)

# parte par e ímpar
x_par = (x + x_neg) / 2
x_impar = (x - x_neg) / 2

# plots
plt.figure(figsize=(10,5))

plt.plot(t, x, label="x(t)")
plt.plot(t, x_par, label="parte par")
plt.plot(t, x_impar, label="parte ímpar")

plt.legend()
plt.grid()
plt.show()
```

<h4 style="color: #008000;">Questão 28</h4>
Considere o sistema:

<p style="text-align: center; font-size: 16px;">
    \( y(t) = x(t - 2) + 2x(t) \)
</p>

```python
# eixo do tempo
t = np.linspace(0, 10, 1000)

# sinal de entrada
x = np.sin(t)

# sinal deslocado
x_shift = np.sin(t - 2)

# saída do sistema
y = np.sin(t-2) + 2*x

# gráfico
plt.plot(t, x, label='x(t)')
plt.plot(t, x_shift, label='x(t-2)')
plt.plot(t, y, label='y(t)')

plt.xlabel('t')
plt.ylabel('Amplitude')

plt.grid()
plt.legend()

plt.show()
```

<h4 style="color: #008000;">Questão 32</h4>
Considere o sistema:

<p style="text-align: center; font-size: 16px;">
    \( y(t) = \int_{-\infty}^{t} x(\tau)\, d\tau \)
</p>

```python
# eixo do tempo
t = np.linspace(-5, 5, 1000)
dt = t[1] - t[0]

# sinal de entrada x(t) (exemplo)
x = np.sin(2*t) * (t >= 0)   # sinal causal (opcional)

# sistema: y(t) = integral de -inf até t
y = np.cumsum(x) * dt

# plots
plt.figure(figsize=(10,6))

plt.subplot(2,1,1)
plt.plot(t, x)
plt.title("Entrada x(t)")
plt.grid()

plt.subplot(2,1,2)
plt.plot(t, y)
plt.title("Saída y(t) = ∫ x(τ)dτ")
plt.grid()

plt.tight_layout()
plt.show()
```

<h4 style="color: #008000;">Questão 34</h4>
Considere o sistema:

<p style="text-align: center; font-size: 16px;">
    \( y(t) = x(t) + x(t-1) \)
</p>

```python
# eixo do tempo
t = np.linspace(0, 10, 1000)

# sinal original
x = np.cos(t)

# sinal atrasado
x_shift = np.sin(t - 1)

# saída do sistema
y = x + x_shift

# gráficos
plt.plot(t, x, label='x(t)')
plt.plot(t, x_shift, label='x(t-1)')
plt.plot(t, y, label='y(t)')

plt.xlabel('t')
plt.ylabel('Amplitude')

plt.grid()
plt.legend()

plt.show()
```
