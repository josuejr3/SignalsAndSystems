<h4>**Questão 2**</h4>

Sendo:
<p>
    \( g(t) = \cos(\omega_1 t)\)
</p>
<p>
    \( h(t) = \cos(\omega_1 t + \theta) \)
</p>
Determine: 
<p>
    \( s(t) = g(t)h(t)\)
</p>
Escolha três valores para θ e esboce os sinais resultantes.

<h4>Para teta = pi</h4>

```python
t = np.linspace(0, 5, 1000) # Cria o eixo do tempo

w = 2
theta = np.pi

g = np.cos(w * t)
h = np.cos(w* t + theta)
s = g*h

plt.plot(t, s)
plt.yticks(np.arange(-1, 0.1, 0.1))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<h4>Para teta = pi/2</h4>

```python
theta = np.pi/2

g = np.cos(w * t)
h = np.cos(w* t + theta)
s = g*h

plt.plot(t, s)
plt.yticks(np.arange(-1, 1, 0.1))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<h4>Para teta = 2pi</h4>

```python
theta = 2 * np.pi

g = np.cos(w * t)
h = np.cos(w* t + theta)
s = g*h

plt.plot(t, s)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<h4>**Questão 4**</h4>

<p>Seja:</p>

<p>
    \( z(t) = e^{j2\pi100t} \)
</p>

<p>Esboce:</p>

<p>
    a) \( x(t) = \Re\{z(t)\} \)
</p>

<p>
    b) \( y(t) = \Im\{z(t)\} \)
</p>

<h4>Gráfico para parte real</h4>

```python
t = np.linspace(0, 0.05, 1000)

# definição do número complexo
z = np.exp(1j * 2 * np.pi * 100 * t)

plt.plot(t, np.real(z))
plt.grid()
plt.show()
```

<h4>Gráfico para parte imaginária</h4>

```python
t = np.linspace(0, 0.05, 1000)

# definição do número complexo
z = np.exp(1j * 2 * np.pi * 100 * t)

plt.plot(t, np.imag(z))
plt.grid()
plt.show()
```

<h4>**Questão 6**</h4>

<p>Sejam:</p>

<p>
    \( k(t) = 2t + 1 \)
</p>

<p>
    \( y(t) = t - 2 \)
</p>

<p>Determine:</p>

<p>
 a)   \( z(t) = k(t) + y(t) \)
</p>

```python
t = np.linspace(0, 5, 1000)
k = (2 * t) + 1
y = t - 2
z = k + y

plt.plot(t, z)
plt.yticks(np.arange(-1, 15, 1))
plt.xlabel('t')
plt.ylabel('z(t)')
plt.grid()
plt.show()
```

<p>
 b)   \( f(t) = k(t)y(t) \)
</p>

```python
t = np.linspace(-4, 5, 1000)
k = (2 * t) + 1
y = t - 2
z = k * y

plt.plot(t, z)
plt.yticks(np.arange(-3, 40, 2))
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.show()
```

<p>
 c)   \( s(t) = k(t-1) \)
</p>

```python
t = np.linspace(0, 5, 1000)
k = (2 * t) + 1
y = t - 2
s = k 

plt.plot(t-1, s)
plt.yticks(np.arange(-3, 10, 2))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<p>
 d)   \( x(t) = y(t+1) \)
</p>

```python
t = np.linspace(0, 5, 1000)
k = (2 * t) + 1
y = t - 2
x = y

plt.plot(t+2, x)
plt.yticks(np.arange(-3, 10, 2))
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid()
plt.show()
```

<p>
 e)   \( j(t) = k(2t+2) \)
</p>

```python
t = np.linspace(0, 5, 1000)
k = (2 * t) + 1
y = t - 2
j = k

plt.plot((2*t)+2, j)
plt.yticks(np.arange(-3, 10, 2))
plt.xlabel('t')
plt.ylabel('j(t)')
plt.grid()
plt.show()
```

<p>
 f)   \( l(t) = y(-t+1) \)
</p>

```python
t = np.linspace(0, 5, 1000)
k = (2 * t) + 1
y = t - 2
l = y

plt.plot((-t+1), l)
plt.yticks(np.arange(-3, 10, 2))
plt.xlabel('t')
plt.ylabel('l(t)')
plt.grid()
plt.show()
```

<h4>**Questão 12**</h4>
Dados os sinais abaixo, determine o valor de pico e o período.

<p>
a)    \( s_1(t) = 10\cos\left(100t + \frac{\pi}{3}\right) + 16\sin\left(150t + \frac{\pi}{6}\right) \)
</p>

```python
t = np.linspace(0, 0.5, 1000)
s1 = 10 * np.cos((100*t) + np.pi/3) + 16 * np.sin((150*t) + np.pi/6)

plt.plot(t, s1)
plt.yticks(np.arange(-26, 28, 2))
plt.xlabel('t')
plt.ylabel('s1(t)')
plt.grid()
plt.show()
```

<p>
b)    \( s_2(n) = 2\cos\left(\frac{\pi n}{2}\right) + 3\cos\left(\frac{2\pi n}{5}\right) \)
</p>

```python
n = np.arange(0, 30)
s2 = 2 * np.cos((np.pi * n)/2) + 3 * np.cos((2*np.pi*n)/5)

# gráfico discreto
plt.stem(n, s2)
plt.xlabel('n')
plt.ylabel('s2(n)')
plt.grid()
plt.show()
```
