<p style="font-size: 16px;">
    a) \( x(n - 1) \)
</p>

```python
# Vetor n
n = np.arange(-2, 8)

# Degrau unitário u(n)
u1 = np.where((n - 1) >= 0, 1, 0)
u5 = np.where((n - 5) >= 0, 1, 0)

# Sinal original
x = u1 - u5

# Sinal deslocado: x(n - 1)
# x(n-1) = u(n-2) - u(n-6)
u2 = np.where((n - 2) >= 0, 1, 0)
u6 = np.where((n - 6) >= 0, 1, 0)

x_shift = u2 - u6

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, x, linefmt='b-', markerfmt='bo', basefmt='k-',
         label='x(n)')

plt.stem(n, x_shift, linefmt='r-', markerfmt='ro', basefmt='k-',
         label='x(n-1)')

plt.title('Sinais x(n) e x(n-1)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.legend()
plt.grid(True)

plt.show()
```

<p style="font-size: 16px;">
    b) \( x(2 - n) \)
</p>

```python
# Vetor n
n = np.arange(-4, 8)

# Sinal transformado:
# x(2 - n) = u(1 - n) - u(-3 - n)

u1 = np.where((1 - n) >= 0, 1, 0)
u2 = np.where((-3 - n) >= 0, 1, 0)

x_transformado = u1 - u2

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, x_transformado,
         linefmt='b-',
         markerfmt='bo',
         basefmt='k-')

plt.title('Sinal x(2 - n)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.grid(True)

plt.show()
```

<h4 style="color: #008000;">Questão 33</h4>

<p style="text-align: center; font-size: 16px;">
    \( y(t) = \dfrac{d}{dt}x(t) \)
</p>

O sistema é:
- Linear
- Não Causal
- Com Memória
- Invariante no Tempo

```python
# 1. Definir o tempo com passos pequenos para maior precisão na derivada
t = np.linspace(-0.5, 4, 3000)
dt = t[1] - t[0] # Passo de tempo (dx)

# 2. Criar a entrada do sistema x(t)
x = np.where(t >= 1, np.sin(2 * np.pi * t) / t, 0.0)

# 3. Calcular a saída y(t) aplicando o operador derivada com np.gradient
y = np.gradient(x, dt)

# 4. Criar o gráfico com dois subplots compartilhando o eixo X
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 7), sharex=True)

# Subplot 1: Entrada do Sistema
ax1.plot(t, x, color='blue', linewidth=2, label='Entrada x(t)')
ax1.axhline(0, color='black', linewidth=0.5)
ax1.axvline(1, color='red', linestyle='--', linewidth=1)
ax1.set_title(r'Sistema Analisado: $y(t) = \frac{d}{dt}x(t)$')
ax1.set_ylabel('Amplitude de Entrada')
ax1.grid(True, linestyle=':')
ax1.legend()

# Subplot 2: Saída do Sistema (Sinal Transformado)
ax2.plot(t, y, color='red', linewidth=2, label='Saída Transformada y(t)')
ax2.axhline(0, color='black', linewidth=0.5)
ax2.axvline(1, color='red', linestyle='--', linewidth=1, label='Salto Abrupto (Derivada -> ∞)')
ax2.set_xlabel('Tempo (t)')
ax2.set_ylabel('Amplitude de Saída')
ax2.set_xlim(-0.5, 4)
ax2.set_ylim(-15, 30) # Limitador para visualizar a oscilação pós-salto
ax2.grid(True, linestyle=':')
ax2.legend()

plt.tight_layout()
plt.show()
```

<h4 style="color: #008000;">Questão 36</h4>
Considere o sistema:

<p style="text-align: center; font-size: 16px;">
    \( m\frac{d^2 y(t)}{dt^2} + b\frac{dy(t)}{dt} + ky(t) = x(t) \)
</p>

O sistema é:
- Linear
- Causal
- Com Memória
- BIBO Estável

```python
from scipy.signal import lti, lsim

# =====================================
# PARÂMETROS DO SISTEMA
# =====================================
m = 1.0
b = 2.0
k = 5.0

# =====================================
# TEMPO
# =====================================
t = np.linspace(0, 10, 1000)

# =====================================
# SINAL ORIGINAL
# =====================================
x = np.sin(2 * np.pi * t)

# =====================================
# SINAL TRANSFORMADO
# (exemplo: amplificação)
# =====================================
x_transformado = 2 * x

# =====================================
# SISTEMA
# G(s) = 1 / (ms² + bs + k)
# =====================================
num = [1]
den = [m, b, k]

sistema = lti(num, den)

# =====================================
# RESPOSTA DO SISTEMA
# =====================================
t_saida, y, _ = lsim(sistema, U=x_transformado, T=t)

# =====================================
# GRÁFICO
# =====================================
plt.figure(figsize=(12,6))

# sinal original
plt.plot(t, x, label='Sinal Original')

# sinal transformado
plt.plot(t, x_transformado, label='Sinal Transformado')

# saída do sistema
plt.plot(t_saida, y, label='Saída do Sistema')

plt.title('Sistema Massa-Mola-Amortecedor')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.show()
```

<p>
    <b>Interpretação Física</b>
</p>

O sistema representa um modelo físico massa-mola-amortecedor. A massa m representa a inércia do sistema, a mola com constante k tende a trazer a massa de volta à posição de equilíbrio e o amortecedor b dissipa energia, reduzindo as oscilações. A entrada x(t) corresponde a uma força aplicada ao sistema e a saída y(t) representa o deslocamento da massa ao longo do tempo.
