```python
t = np.linspace(0, 2*np.pi, 1000)
theta_b = np.pi/2

g_b = np.cos(3*t)
h_b = np.cos(3*t + theta_b)

x_b = g_b + h_b
# Criando os gráficos
plt.figure(figsize=(12,10))
plt.subplot(3,1,2)
plt.plot(t, x_b, linewidth=2)
plt.title('b) θ = π/2')
plt.ylabel('x(t)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

</p>

```python
# Tempo
t = np.linspace(0, 2*np.pi, 1000)

# θ = π
theta = np.pi

# Sinais
g = np.cos(3*t)
h = np.cos(3*t + theta)

# Soma dos sinais
x = g + h

# Eliminando erros numéricos
x = np.round(x, 10)

# Plotando
plt.figure(figsize=(10,5))

plt.plot(t, x, linewidth=2)

plt.title(r'Sinal $x(t)=\cos(3t)+\cos(3t+\pi)$')
plt.xlabel('t')
plt.ylabel('x(t)')

plt.grid(True)

plt.show()
```

<h4 style="color: #008000;">Questão 9</h4>

Considere:

<p style="text-align: center; font-size: 16px;">
    \( x(n) = u(n - 1) - u(n - 5) \)
</p>
Esboce:

<p style="font-size: 16px;">
    a) \( x(n - 1) \)
</p>

```python
# Vetor n
n = np.arange(-2, 8)

# Degrau unitário u(n)
u1 = np.where((n - 1) >= 0, 1, 0)
u5 = np.where((n - 5) >= 0, 1, 0)

# Sinal original
x = u1 - u5

# Sinal deslocado: x(n - 1)
# x(n-1) = u(n-2) - u(n-6)
u2 = np.where((n - 2) >= 0, 1, 0)
u6 = np.where((n - 6) >= 0, 1, 0)

x_shift = u2 - u6

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, x, linefmt='b-', markerfmt='bo', basefmt='k-',
         label='x(n)')

plt.stem(n, x_shift, linefmt='r-', markerfmt='ro', basefmt='k-',
         label='x(n-1)')

plt.title('Sinais x(n) e x(n-1)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.legend()
plt.grid(True)

plt.show()
```

<p style="font-size: 16px;">
    b) \( x(2 - n) \)
</p>

```python
# Vetor n
n = np.arange(-4, 8)

# Sinal transformado:
# x(2 - n) = u(1 - n) - u(-3 - n)

u1 = np.where((1 - n) >= 0, 1, 0)
u2 = np.where((-3 - n) >= 0, 1, 0)

x_transformado = u1 - u2

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, x_transformado,
         linefmt='b-',
         markerfmt='bo',
         basefmt='k-')

plt.title('Sinal x(2 - n)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.grid(True)

plt.show()
```

<p style="font-size: 16px;">
    c) \( x(2n + 1) \)
</p>

```python
# Vetor n
n = np.arange(-5, 5)

# Sinal transformado:
# x(2n + 1) = u(2n) - u(2n - 4)

u1 = np.where((2*n) >= 0, 1, 0)
u2 = np.where((2*n - 4) >= 0, 1, 0)

x_transformado = u1 - u2

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, x_transformado,
         linefmt='b-',
         markerfmt='bo',
         basefmt='k-')

plt.title('Sinal x(2n + 1)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.grid(True)

plt.show()
```

<p style="font-size: 16px;">
    d) \( x(n)u(2 - n) \)
</p>

```python
# Vetor n
n = np.arange(-2, 8)

# Sinal original:
# x(n) = u(n-1) - u(n-5)

u1 = np.where((n - 1) >= 0, 1, 0)
u5 = np.where((n - 5) >= 0, 1, 0)

x = u1 - u5

# Degrau u(2 - n)
u = np.where((2 - n) >= 0, 1, 0)

# Sinal resultante
y = x * u

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, y,
         linefmt='b-',
         markerfmt='bo',
         basefmt='k-')

plt.title('Sinal x(n)u(2 - n)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.grid(True)

plt.show()
```

<p style="font-size: 16px;">
    e) \( x(n - 1)\delta(n - 3) \)
</p>

```python
# Vetor n
n = np.arange(-2, 8)

# x(n-1) = u(n-2) - u(n-6)

u2 = np.where((n - 2) >= 0, 1, 0)
u6 = np.where((n - 6) >= 0, 1, 0)

x_shift = u2 - u6

# Delta de Kronecker δ(n-3)
delta = np.where(n == 3, 1, 0)

# Sinal resultante
y = x_shift * delta

# Plotando
plt.figure(figsize=(10,5))

plt.stem(n, y,
         linefmt='b-',
         markerfmt='bo',
         basefmt='k-')

plt.title('Sinal x(n-1)δ(n-3)')
plt.xlabel('n')
plt.ylabel('Amplitude')

plt.grid(True)

plt.show()
```

<h4 style="color: #008000;">Questão 32</h4>
Considere o sistema:

<p style="text-align: center; font-size: 16px;">
    \( y(t) = \int_{-\infty}^{t} x(\tau)\, d\tau \)
</p>

```python
# eixo do tempo
t = np.linspace(-5, 5, 1000)
dt = t[1] - t[0]

# sinal de entrada x(t) (exemplo)
x = np.sin(2*t) * (t >= 0)   # sinal causal (opcional)

# sistema: y(t) = integral de -inf até t
y = np.cumsum(x) * dt

# plots
plt.figure(figsize=(10,6))

plt.subplot(2,1,1)
plt.plot(t, x)
plt.title("Entrada x(t)")
plt.grid()

plt.subplot(2,1,2)
plt.plot(t, y)
plt.title("Saída y(t) = ∫ x(τ)dτ")
plt.grid()

plt.tight_layout()
plt.show()
```

<h4 style="color: #008000;">Questão 36</h4>
Considere o sistema:

<p style="text-align: center; font-size: 16px;">
    \( m\frac{d^2 y(t)}{dt^2} + b\frac{dy(t)}{dt} + ky(t) = x(t) \)
</p>

```python
from scipy.signal import lti, lsim

# =====================================
# PARÂMETROS DO SISTEMA
# =====================================
m = 1.0
b = 2.0
k = 5.0

# =====================================
# TEMPO
# =====================================
t = np.linspace(0, 10, 1000)

# =====================================
# SINAL ORIGINAL
# =====================================
x = np.sin(2 * np.pi * t)

# =====================================
# SINAL TRANSFORMADO
# (exemplo: amplificação)
# =====================================
x_transformado = 2 * x

# =====================================
# SISTEMA
# G(s) = 1 / (ms² + bs + k)
# =====================================
num = [1]
den = [m, b, k]

sistema = lti(num, den)

# =====================================
# RESPOSTA DO SISTEMA
# =====================================
t_saida, y, _ = lsim(sistema, U=x_transformado, T=t)

# =====================================
# GRÁFICO
# =====================================
plt.figure(figsize=(12,6))

# sinal original
plt.plot(t, x, label='Sinal Original')

# sinal transformado
plt.plot(t, x_transformado, label='Sinal Transformado')

# saída do sistema
plt.plot(t_saida, y, label='Saída do Sistema')

plt.title('Sistema Massa-Mola-Amortecedor')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.show()
```

<p>
    <b>Interpretação Física</b>
</p>

O sistema se trata de um sistema massa mola não amortecido, em que o valor K é a constante de elasticidade da mola. b é o coeficiente de amortecimento e o valor de m representa a massa.
